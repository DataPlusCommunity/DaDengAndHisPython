### Êï£ÁÇπÂõæ
- üç≠Âü∫Á°ÄÁ§∫‰æãÔºõ

```python
data_x = [random.randint(1, 100) for _ in range(1000)]
data_y = [random.randint(1, 100) for _ in range(1000)]
          
plt.figure(figsize=(8, 8))
plt.scatter(data_x, data_y)
plt.show()
```

- üç®ËÆæÁΩÆÊï£ÁÇπÈ¢úËâ≤Ôºõ
- üç©ËÆæÁΩÆÊï£ÁÇπÂ§ßÂ∞èÔºõ
- ü•§ÈÄèÊòéÂ∫¶ËÆæÁΩÆÔºõ

```python
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2 

plt.figure(figsize=(8, 8))
plt.scatter(x, y, 
            # ÂõæÂΩ¢Â§ßÂ∞è
            s=area, 
            # È¢úËâ≤
            c=colors, 
            # ÈÄèÊòéÂ∫¶
            alpha=0.3)
plt.show()
```

### ÂõæË°®Âä®ÊÄÅÂà∑Êñ∞

#### ü•´‰∫§‰∫íÊ®°Âºè
- Êàë‰ª¨Âú®ËÆ≠ÁªÉÊ®°ÂûãÁöÑÊó∂ÂÄôÔºåÂèØËÉΩÈúÄË¶ÅÂéª‰∏çÂÅúÂéªÊõ¥Êñ∞ÂõæË°®Êï∞ÊçÆÔºõ

```python
# jupyter‰∏≠Âä®ÊÄÅÊòæÁ§∫ÈúÄË¶Å
from IPython import display

x, y1, y2 = [], [], []
plt.ion()    # ‰∫§‰∫íÊ®°ÂºèÊâìÂºÄ
for i in np.linspace(0,20,500):
    x.append(i)
    y1.append(np.cos(i))
    y2.append(np.sin(i))
    plt.figure(figsize=(12, 2))
    plt.clf()   
    plt.ylim(-1, 1)    # ËÆæÁΩÆyËΩ¥ÂàªÂ∫¶ÂÄºÊúÄÂ§ßÊúÄÂ∞èÂÄº
    plt.plot(x, y1, '--r')
    plt.plot(x, y2, 'g')
    plt.pause(0.1)
    display.clear_output(wait=True)    # Ê∏ÖÈô§ËæìÂá∫
plt.ioff()   # ÂÖ≥Èó≠‰∫§‰∫íÊ®°Âºè
```

### Â≠êÂõæ/ÂõæË°®È£éÊ†º
- üç∞Â§öÂ≠êÂõæÔºõ
- üçßÂõæË°®È£éÊ†ºËÆæÁΩÆÔºõ

```python
# ÂèØÁî®È£éÊ†º
style_list = plt.style.available

# Êï∞ÊçÆ
cate = ['Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'Meizu']
data1 = [89, 45, 56, 74, 23, 21]
data2 = [56, 77, 93, 68, 45, 67]
x_location = np.arange(len(cate))

plt.figure(figsize=(20, 40))
for i, style in enumerate(style_list):
    
    with plt.style.context(style):  # ËÆæÁΩÆÂõæË°®È£éÊ†º     
        plt.subplot(7, 4, i+1)
        plt.bar(x_location, data1, label='Á∫ø‰∏ã', width=0.3)
        plt.bar(x_location+0.3, data2, label='Á∫ø‰∏ä', width=0.3)

        # ‰∏∫XËΩ¥ËÆæÁΩÆÂàªÂ∫¶ÂÄº
        plt.xticks(x_location+0.15, cate)
        plt.title(style,
                 fontdict={'fontsize':20,'color':'red'})
plt.show()
```
